## Performing Analysis 

```python
#Matches conducted on each month
datcopy = resdat.copy()
datcopy['Date'] = pd.to_datetime(datcopy['Date'])
datcopy.set_index('Date', inplace=True)  # Set the 'Date' column as the index
mpm = datcopy.resample('M').size()
print("Matches conducted each month:",mpm)

plt.figure(figsize =(15,6))
mpm.plot(kind='bar',color = sns.color_palette('Set2'))
plt.xlabel('Months')
plt.ylabel('No. of Matches')
plt.title('Matches conducted per Month')
plt.xticks(rotation = 90)
plt.show()


```

```python
#No of matches conducted on each time
met = resdat.groupby('Time').size()
sns.set(style='whitegrid')
plt.figure(figsize =(20,8))
met.plot(kind='bar',color = sns.color_palette('magma'))
plt.xlabel('Time in GMT')
plt.ylabel('No. of Matches')
plt.title('Matches conducted on each time')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Goals scored in first half and second half
goals_first_half = resdat['HTHG'].sum() + resdat['HTAG'].sum()
goals_second_half = (resdat['FTHG'].sum()+resdat['FTAG'].sum()) - goals_first_half

print("Goals Scored in First Half:", goals_first_half)
print("Goals Scored in Second Half:", goals_second_half)

plt.figure(figsize=(10,6))
data = [goals_first_half, goals_second_half]
labels = ['First half Goals', 'Second half Goals']
color = sns.color_palette("hls", 4)
plt.pie(data, labels = labels, colors = color, autopct = "%.f%%")
plt.show()
```

```python
#Goals scored by home teams and away teams in first half and second half
firsthalf_home = resdat['HTHG'].sum()
firsthalf_away = resdat['HTAG'].sum()

secondhalf_home = resdat['FTHG'].sum() - firsthalf_home
secondhalf_away = resdat['FTAG'].sum() - firsthalf_away

print("Goals scored by home team during first half:", firsthalf_home)
print("Goals scored by away team during first half:", firsthalf_away)

print("Goals scored by home team during second half:", secondhalf_home)
print("Goals scored by away team during second half:", secondhalf_away) 

half_labels = ['First Half', 'Second Half']
home_goals = [firsthalf_home, secondhalf_home]
away_goals = [firsthalf_away, secondhalf_away]

# Create positions for the bars
x = np.arange(len(half_labels))
width = 0.25  # Width of the bars

# Create the grouped bar plot
fig, ax = plt.subplots(figsize=(6,6))
rects1 = ax.bar(x - width/2, home_goals, width, label='Home Team')
rects2 = ax.bar(x + width/2, away_goals, width, label='Away Team')

# Customize the plot
ax.set_ylabel('Goals')
ax.set_title('Goals Scored by Home and Away Teams')
ax.set_xticks(x)
ax.set_xticklabels(half_labels)
ax.legend()
plt.show()
```

```python
#Full Time result when hometeam is leading at HT vs FT result when away team is leading at HT
home_leading = resdat[resdat['HTHG']>resdat['HTAG']]
away_leading = resdat[resdat['HTAG']>resdat['HTHG']]
draw_halftime = resdat[resdat['HTHG'] == resdat['HTAG']]

FTR_home_leading = home_leading['FT Result'].value_counts()
FTR_away_leading = away_leading['FT Result'].value_counts()
FTR_draw = draw_halftime['FT Result'].value_counts()

print("Home Team is leading at Half Time")
print("Home Team win:", FTR_home_leading[0])
print("Away Team win:", FTR_home_leading[1])
print("Match tied:", FTR_home_leading[2])

print("\nAway Team is leading at Half Time")
print("Home Team win:", FTR_away_leading[0])
print("Away Team win:", FTR_away_leading[1])
print("Match tied:", FTR_away_leading[2])

print("\nMatch is tied at Half Time")
print("Home Team win:", FTR_draw[0])
print("Away Team win:", FTR_draw[1])
print("Match tied:", FTR_draw[2])

color_palette = sns.color_palette("husl", 3)

plt.figure(figsize=(15, 6))

# Plot for home team leading at HT
plt.subplot(1, 3, 1)
FTR_home_leading.plot(kind='bar', color=color_palette, alpha=0.7)
plt.xlabel("Full Time Result")
plt.ylabel("Count")
plt.title("FTR when Home Team is Leading at Halftime")

# Plot for away team leading at HT
plt.subplot(1, 3, 2)
FTR_away_leading.plot(kind='bar', color=color_palette, alpha=0.7)
plt.xlabel("Full Time Result")
plt.ylabel("Count")
plt.title("FTR when Away Team is Leading at Halftime")

#Plot for Draw during HalfTime
plt.subplot(1, 3, 3)
FTR_draw.plot(kind='bar', color=color_palette, alpha=0.7)
plt.xlabel("Full Time Result")
plt.ylabel("Count")
plt.title("FTR when both teams are tied")

plt.tight_layout()
plt.show()
```

```python
#Matches won by home team and away team
Home_team = resdat[resdat['FT Result'] == 'H']
Away_team = resdat[resdat['FT Result'] == 'A']
no_goals = resdat[resdat['FT Result'] == 'D']

HT_win = Home_team['FT Result'].count()
AT_win = Away_team['FT Result'].count()
match_tied = no_goals['FT Result'].count()

print("Matches won by Home Team:", HT_win, end="")
print("\tMatches won by Away Team:", AT_win, end="")
print("\tMatches Tied:", match_tied)

#plotting graph
data = [HT_win, AT_win, match_tied]
labels = ["Home Team","Away Team","Tied"]

bp = plt.bar(labels, data, color = ["red","blue","black"], alpha = 0.8, width = 0.8)
plt.xlabel("Match result")
plt.ylabel("No. of matches")
plt.title("Matches won by Home Team and Away Team")

for bars,label in zip(bp, labels):
    bars.set_label(label)
plt.legend()
plt.show()
```

```python
#Comparison of total shots vs total shorts on target
total_shots = resdat['Home Shots'].sum() + resdat['Away Shots'].sum()
total_shots_on_target = resdat['Home Shots on Target'].sum() + resdat['Away Shots on Target'].sum()

print("Total no. of Shots this season:",total_shots)
print("Shots on Target this season:",total_shots_on_target)

plt.figure(figsize=(8,4))
data = [total_shots, total_shots_on_target]
labels = ['Shots', 'Shots on Target']
color = sns.color_palette("tab10")
plt.pie(data, labels = labels, colors = color, autopct = "%.f%%")
plt.show()
```

```python
#shots on target vs goals scored
goals_scored = goals_first_half + goals_second_half
print("Total No. of Goals Scored:", goals_scored)

#plot the graph
plt.figure(figsize=(8,4))
data = [total_shots_on_target, goals_scored]
labels = ["Shots on Target", "Goals"]
color = sns.color_palette("husl",3)
plt.pie(data, labels = labels,colors = color, autopct = "%.f%%")
plt.show()
```

```python
#home team corners vs away team corners
total_corners = (resdat['Home Corners'] + resdat['Away Corners']).sum()
ht_corners = resdat['Home Corners'].sum()
at_corners = resdat['Away Corners'].sum()

plt.bar(['Total Corners'], [total_corners], color='limegreen')
plt.bar(['Home Team Corners'], [ht_corners], color='royalblue')
plt.bar(['Away Team Corners'], [at_corners], color='tomato')

plt.xlabel("Corners comparison")
plt.ylabel("Number of Corners")
plt.title("Corners Taken")
labels = ['Total Corners', 'Home Team Corners', 'Away Team Corners']
plt.legend(labels)
plt.show()

print("Total corners taken:", total_corners)
print("Corners taken by Home Team:", ht_corners)
print("Corners taken by Away Team:", at_corners)
```

```python
#Heatmap for the results based on corners
result_counts = resdat.groupby(['Home Corners', 'Away Corners', 'FT Result']).size().unstack(fill_value=0)

a = ['A','H','D']
for i in a:
    pivot_table = result_counts.pivot_table(index='Away Corners', columns='Home Corners', values=i, fill_value=0)
    title = f"FT Result '{i}' Counts Based on Home and Away Team Corners"

    plt.figure(figsize=(8, 6))
    sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu')
    plt.title(title)
    plt.show()
```

```python
#Match ftr based on referees
match_res = resdat.groupby(['Referee','FT Result']).size().unstack()

match_res.plot(kind='bar', stacked=False, figsize=(12, 6))

plt.xlabel("Referee")
plt.ylabel("Number of Matches")
plt.title("Match Results Based on Referees")
plt.legend(title="FT Result", loc='upper right')
plt.show()

match_res
```

```python
#Yellow cards and red cards given by referees
referee_stats = resdat.groupby('Referee')[['Home Yellow Cards', 'Away Yellow Cards', 'Home Red Cards', 'Away Red Cards']].sum()
referee_stats["Yellow cards given"] = referee_stats['Home Yellow Cards'] + referee_stats['Away Yellow Cards']
referee_stats["Red cards given"] = referee_stats['Home Red Cards'] + referee_stats['Away Red Cards']
#referee_stats.drop([['Home Yellow Cards', ' Away Yellow Cards', 'Home Red Cards', 'Away Red Cards']])

referee_stats[['Yellow cards given', 'Red cards given']].plot(kind='bar', figsize=(10, 6))

plt.xlabel("Referee")
plt.ylabel("Number of Cards Given")
plt.title("Yellow and Red Cards Given by Referees")
plt.legend(title="Card Type", loc='upper right')
plt.show()


referee_stats[['Yellow cards given', 'Red cards given']]
```

```python
#fouls given and their comparison
total_fouls = (resdat['Home Fouls'] + resdat['Away Fouls']).sum()
home_fouls = resdat['Home Fouls'].sum()
away_fouls = resdat['Away Fouls'].sum()

plt.bar(['Total Fouls'], [total_fouls], color='limegreen')
plt.bar(['Home Team Fouls'], [home_fouls], color='royalblue')
plt.bar(['Away Team Fouls'], [away_fouls], color='tomato')

for i, v in enumerate([total_fouls, home_fouls, away_fouls]):
    plt.text(i, v, str(v), ha='center', va='bottom', fontsize=10)

plt.xlabel("Fouls comparison")
plt.ylabel("Number of Fouls")
plt.title("Fouls Given")
labels = ['Total Fouls', 'Home Team Fouls', 'Away Team Fouls']
plt.legend(labels)
plt.show()

print("Total Fouls Given:", total_fouls)
print("Fouls given for home team:", home_fouls)
print("Fouls given for away team:", away_fouls)
```
